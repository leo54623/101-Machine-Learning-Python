```python
# ===========================
# Mapa de prescripción del homicidio en el mundo
# ===========================

import matplotlib.pyplot as plt
import geopandas as gpd
import matplotlib.patches as mpatches
from geodatasets import get_path # <-- Añadí esta importación

# ---------------------------
# Datos de prescripción (0 = prescribe, 1 = no prescribe)
# ---------------------------
data_prescripcion = {
    "Argentina": 0,
    "Spain": 0,
    "Italy": 0,
    "France": 0,
    "Mexico": 0,
    "Chile": 0,
    "Brazil": 0,
    "Colombia": 0,
    "United States": 1,
    "United Kingdom": 1,
    "Canada": 1,
    "Germany": 1,
    "Australia": 1,
}

# ---------------------------
# Colores
# ---------------------------
color_prescribe = "#4575b4"    # Azul
color_no_prescribe = "#d73027" # Rojo

# ---------------------------
# Cargar mapa mundial
# ---------------------------
import geopandas as gpd
from geodatasets import get_path # <-- Añade esta importación

# Carga el mapa usando la nueva biblioteca
# Esta es la forma correcta de cargar el mapa ahora
world = gpd.read_file(r"C:\Users\leomendez\OneDrive\00. Industria 4_0\110m_cultural\ne_110m_admin_0_countries.shp")

# Agregar columna con valores de prescripción
world["prescripcion"] = world["NAME"].map(data_prescripcion)

# Asignar colores según la prescripción
world["color"] = world["prescripcion"].map({0: color_prescribe, 1: color_no_prescribe})

# ---------------------------
# Graficar
# ---------------------------
# fig, ax = plt.subplots(figsize=(15, 8))


world["color"] = world["prescripcion"].map({0: "#4575b4", 1: "#d73027"}).fillna("#cccccc")
world.plot(color=world["color"], edgecolor="black", linewidth=0.3)


# Leyenda
patch_no = mpatches.Patch(color=color_no_prescribe, label="No prescribe")
patch_si = mpatches.Patch(color=color_prescribe, label="Sí prescribe")
plt.legend(handles=[patch_no, patch_si], loc='lower left')

# Título
plt.title("Prescripción del homicidio en el mundo", fontsize=18, fontweight="bold")
plt.axis("off")

plt.show()
```
