# Texto en negrita y cursiva
Puede utilizar etiquetas br y los simbolos <> , '**' es decir 'doble asterisco' o '__' es decir 'doble guión bajo' para obtener texto en negrita con la siguiente sintaxis.

Resultado :

<b> Este es un texto en negrita 1 </b>

** Este es un texto en negrita 2 ** (Este como que no funciona)

__ Este es un texto en negrita 3 __ (Y este tampoco)

Puede utilizar etiquetas <i> , '*' (es decir, un asterisco simple) o '_' (es decir, un guión bajo simple) para obtener el texto en cursiva para la siguiente sintaxis. No deje un espacio entre el asterisco y el texto porque cuenta como una viñeta.

Resultado :

* Este es un texto en cursiva 2 *
* _ Este es un texto en cursiva 3 _
